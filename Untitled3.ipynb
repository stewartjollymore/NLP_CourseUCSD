```python
#Problem #1

import numpy as np
from scipy import linalg as lg

a = np.array([[2,0],[0,5]])
values, vector = lg.eig(a)

print(values)
print()
print(vector)
print()

b = np.array([[4,1,6],[0,2,3],[0,0,9]])
values, vector = lg.eig(b)

print(values)
print()
print(vector)
print()

c = np.array([[1,4,0,0],[0,2,0,0],[5,3,1,1],[4,7,2,2]])
values, vector = lg.eig(c)

print(values)
print()
print(vector)
```

above for both diagonal and trangular matrixes we see that the eigenvalues are just diagonals, this is a therom in LA. Furthermore, for c) we have a n-1 eighen values since the span of the row space is one less than the dimension of the row space, in otherwords we have lineraly dependent rows.

Lastly, the definition of eignvalue requires `dim(A)=nxn`. So only square matrices have eigenvalues.

```python
#Problem #4

mat1 = np.array([[3,1,1],[-1,3,1]])

V, S, U = lg.svd(mat1)

sig = np.diag(S)
z = np.zeros((2,1))

S = np.hstack((sig,z))
print(np.dot(np.dot(V,S),U).round())

mat2 = np.array([[1,1],[0,1],[-1,1]])

V, S, U = lg.svd(mat2)

sig = np.diag(S)
z = np.zeros((1,2))

S = np.vstack((sig, z))

print()
print(np.dot(np.dot(V,S),U).round())
```

```python
#Problem #5

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD

corpus = ["runs",
          "runs pitcher innings",
          "pitcher",
          "touchdown",
          "quarterback",
          "touchdown quarterback"]

vectorizer = CountVectorizer()

BOW = vectorizer.fit_transform(corpus)
```

```python
from sklearn import metrics

bow_mat = BOW.toarray()

l2norm = np.sqrt((bow_mat*bow_mat).sum(axis = 1))


bow_normd = bow_mat/l2norm.reshape((6,1))

print("Cosine Sim")
print(metrics.pairwise.cosine_similarity(bow_normd,bow_normd).round(4))
print()
print("Cosine Dis")
print(metrics.pairwise.cosine_distances(bow_normd,bow_normd).round(4))
```

```python
dtm = BOW.todense()
tdm = BOW.todense().T

svd = TruncatedSVD(n_components= 2)
lsa_dtm = svd.fit_transform(np.asarray(dtm))
print(lsa_dtm.round(4))
print()

lsa_tdm = svd.fit_transform(np.asarray(tdm))
print(lsa_tdm.round(4))
```

```python
import matplotlib.pyplot as plt

plt.scatter(lsa_dtm[:,0], lsa_dtm[:,1])
plt.xlabel("Truncated DTM/Topic-1 Values")
plt.ylabel("Truncated DTM/Topic-2 Values")
```

```python
print(metrics.pairwise.cosine_similarity(lsa_dtm, lsa_dtm).round(4))
print()
print(metrics.pairwise.cosine_distances(lsa_dtm, lsa_dtm).round(4))
```

```python
import pandas as pd

encd_mat = pd.DataFrame(lsa_dtm.round(4).T,
                        index = ["topic1", "topic2"]).T
encd_mat["docs"] = corpus

display(encd_mat[["docs","topic1","topic2"]])
```

```python
encd_mat = pd.DataFrame(lsa_tdm.round(4).T,
                        index = ["topic1", "topic2"]).T

encd_mat["terms"] = vectorizer.get_feature_names_out()


display(encd_mat[["terms", "topic1", "topic2"]])
```
